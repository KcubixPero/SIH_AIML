```python
import pandas as pd
from sklearn.preprocessing import MultiLabelBinarizer
from scipy import sparse

# Load CSV
df = pd.read_csv("k3.csv")
df.drop_duplicates(subset=['Job Title'], inplace=True)
# Split skills
df['Skills'] = df['Skills'].str.split(',').apply(lambda x: [s.strip() for s in x])

# Use MultiLabelBinarizer for one-hot encoding
mlb = MultiLabelBinarizer(sparse_output=True)
skill_matrix = mlb.fit_transform(df['Skills'])

# skill_matrix is now a sparse matrix (memory efficient)
print(f"Shape: {skill_matrix.shape}, Type: {type(skill_matrix)}")

# Save to disk
sparse.save_npz("skills_sparse.npz", skill_matrix)
pd.DataFrame(mlb.classes_, columns=["Skill"]).to_csv("skills_columns.csv", index=False)
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Load CSV
df = pd.read_csv("k3_ohe_all_skills.csv")

# Candidate skills
candidate_skills = ['ms-Word', 'English proficiency (spoken)']

# Flatten job skills into a single string per role
def get_job_skills_text(row, skill_columns):
    skills = []
    for col in skill_columns:
        val = row[col]
        if isinstance(val, str):
            skills_in_cell = [s.strip() for s in val.strip("[]").replace("'", "").split(',') if s.strip()]
            skills.extend(skills_in_cell)
        elif isinstance(val, (int, float)) and val == 1:
            skills.append(col)
    return " ".join(skills).lower()

# Identify skill columns
skill_columns = [col for col in df.columns if col not in ['Job Title', 'Skills']]

# Create job skill texts
df['skill_text'] = df.apply(lambda row: get_job_skills_text(row, skill_columns), axis=1)

# Candidate text
candidate_text = " ".join(candidate_skills).lower()

# TF-IDF vectorization
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(df['skill_text'].tolist() + [candidate_text])

# Cosine similarity
cos_sim = cosine_similarity(tfidf_matrix[-1], tfidf_matrix[:-1])[0]

# Add similarity to df
df['similarity'] = cos_sim

# Sort top matches
top_df = df.sort_values('similarity', ascending=False).head(5)

# Print matched/missing skills
for _, row in top_df.iterrows():
    job_skills = set(row['skill_text'].split())
    candidate_set = set([s.lower() for s in candidate_skills])
    matched = candidate_set.intersection(job_skills)
    missing = job_skills - candidate_set
    print(f"\nRole: {row['Job Title']}")
    print(f"Similarity Score: {row['similarity']:.2f}")
    print(f"Matched : {matched}")
    print(f"Missing : {missing}")
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from scipy import sparse
import numpy as np

# ------------------------
# Load original job data
# ------------------------
df_jobs = pd.read_csv("k3.csv")  # Must have 'Job Title' column
df_jobs.drop_duplicates(subset=['Job Title'], inplace=True)
# ------------------------
# Load sparse skill matrix
# ------------------------
skill_matrix = sparse.load_npz("skills_sparse.npz")  # shape: (num_jobs, num_skills)
skill_columns = pd.read_csv("skills_columns.csv")['Skill'].tolist()  # list of skill names

# ------------------------
# Candidate skills
# ------------------------
candidate_skills = ['ms-Word', 'English proficiency (spoken)']

# ------------------------
# Convert candidate skills to TF-IDF style vector (matching job text)
# ------------------------
# Flatten job skills into a single string per role
def get_job_skills_text(row):
    skills = row['Skills']
    if isinstance(skills, str):
        skills_in_cell = [s.strip() for s in skills.strip("[]").replace("'", "").split(',') if s.strip()]
        return " ".join(skills_in_cell).lower()
    return ""

df_jobs['skill_text'] = df_jobs.apply(get_job_skills_text, axis=1)

# Candidate text
candidate_text = " ".join(candidate_skills).lower()

# TF-IDF vectorization (job + candidate)
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(df_jobs['skill_text'].tolist() + [candidate_text])

# Cosine similarity
cos_sim = cosine_similarity(tfidf_matrix[-1], tfidf_matrix[:-1])[0]

# Add similarity to df_jobs
df_jobs['similarity'] = cos_sim

# ------------------------
# Get top matches
# ------------------------
top_df = df_jobs.sort_values('similarity', ascending=False).head(5)

# Print matched/missing skills
for _, row in top_df.iterrows():
    job_skills = set(row['skill_text'].split())
    candidate_set = set([s.lower() for s in candidate_skills])
    matched = candidate_set.intersection(job_skills)
    missing = job_skills - candidate_set
    print(f"\nRole: {row['Job Title']}")
    print(f"Similarity Score: {row['similarity']:.2f}")
    print(f"Matched : {matched}")
    print(f"Missing : {missing}")
```
